```python
import cv2 as cv
```

```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np


bottom_left = cv.imread('ImageCollage Dataset/bottom_left.jpg')
top_left = cv.imread('ImageCollage Dataset/top_left.jpg')
bottom_right = cv.imread('ImageCollage Dataset/bottom_right.jpg')
top_right = cv.imread('ImageCollage Dataset/top_right.jpg')
center = cv.imread('ImageCollage Dataset/center.jpeg')

bottom_left_ = cv.cvtColor(bottom_left,cv.COLOR_BGR2RGB)
top_left_ = cv.cvtColor(top_left,cv.COLOR_BGR2RGB)
bottom_right_ = cv.cvtColor(bottom_right,cv.COLOR_BGR2RGB)
top_right_ = cv.cvtColor(top_right,cv.COLOR_BGR2RGB)
center_ = cv.cvtColor(center,cv.COLOR_BGR2RGB)
```

```python

sample = cv.imread('ImageCollage Dataset/sample_image.jpg')
sample = cv.cvtColor(sample,cv.COLOR_BGR2RGB)

plt.imshow(sample)
plt.axis('on')
plt.show()
```

```python
bl = cv.resize(bottom_left_,(200,200))
tl = cv.resize(top_left_,(200,200))
br = cv.resize(bottom_right_,(200,200))
tr = cv.resize(top_right_,(200,200))
ce = cv.resize(center_,(100,100))
```

```python
"""print(bl.shape)
plt.imshow(bl)
plt.axis('off')
plt.show() """


bl = np.array([bl])
tl = np.array([tl])
br = np.array([br])
tr = np.array([tr])
ce = np.array([ce])
```

```python
base = np.zeros((430,430,3),dtype=int)
print(base)
plt.imshow(base)
```

```python
base.shape
```

```python
base[10:210,10:210,:] = tl
base[220:420,220:420,:] = br
base[10:210,220:420,:] = tr
base[220:420,10:210,:] = bl


base[155:275,155:275,:] = 0
base[165:265,165:265,:] = ce





```

```python
plt.imshow(base)

plt.axis('off')



```

```python
collage = np.reshape(base,(base.shape[0]*base.shape[1],3))
```

```python
collage.shape
```

```python
df = pd.DataFrame(data = collage)
```

```python
df.to_csv('PhotoCollage.csv',index = False,header=['r','g','b'])
```
